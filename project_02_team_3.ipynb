### 1. Deep neural network model

**Fundamentos de las redes neuronales**

Las redes neuronales artificiales (RNA) son modelos matemáticos y computacionales que emulan el comportamiento de las neuronas biológicas del cerebro. Inspiradas en la manera en que el cerebro procesa información, las RNA están formadas por unidades de procesamiento básicas llamadas neuronas, las cuales se organizan en capas y se interconectan a través de sinapsis (o conexiones) ponderadas por valores numéricos llamados pesos. Esta estructura permite que las redes neuronales aprendan patrones complejos y realicen diversas tareas, como clasificación y regresión.

Dentro de este contexto, una red neuronal profunda es un tipo de red neuronal artifical con múltipes capas ocultas que aportan un mayor nivel de complejidad y uso intensivo de recursos comparado a las redes neuronales convencionales que usan una o dos capas ocultas. Estas redes neuronales profundas son usadas para distintas aplicaciones y servicios, principalmente en la automatización de tareas analíticas o físicas para limitar la intervención humana, su arquitectura suele ser basada en GPU para entrenamientos rápidos, ya que a medida que se añaden neuronas y capas la red necesita mayor cantidad de recursos así como tiempo para entrenar. También pueden emplear el aprendizaje no supervisado lo que permite extraer cacracterísticas, funciones y relaciones de las variables de entradas para aumentar la precisión. (IBM, 2024)



 **Componentes de una red neuronal profunda**

1.	**Entradas:** Las entradas representan las características o variables predictoras del conjunto de datos. Cada entrada tiene un valor numérico, que puede ser una característica extraída de un conjunto de datos como, por ejemplo, el tamaño de un objeto, el color, o cualquier otro atributo relevante para la tarea que se desea resolver.
2.	**Pesos:** A cada entrada se le asigna un peso que determina su relevancia en el cálculo de la salida de la neurona. Un peso alto significa que la entrada tiene una mayor influencia en la salida, mientras que un peso bajo disminuye su impacto. El objetivo del entrenamiento de la red es ajustar estos pesos de manera que la red pueda tomar decisiones correctas al clasificar nuevas entradas.
3.	**Sesgo:** El sesgo es un valor adicional que se suma a la suma ponderada de las entradas antes de aplicar la función de activación. El sesgo permite que la función de activación se desplace y facilita la activación de la neurona incluso cuando todas las entradas tienen valores bajos. En esencia, el sesgo ayuda a que la red sea más flexible y se ajuste mejor a los datos.
4.	**Función de activación:** Después de calcular la entrada neta, esta pasa a través de una función de activación en donde la salida de la neurona se propaga a las siguientes capas o a la salida final de la red, lo que introduce una no linealidad al modelo permitiendo aprender patrones complejos entre los datos. De las más utilizadas se encuentra la sigmoidal, tanh y ReLU.
5. **Capas:** Una capa es un componente que aplica una transformación matemática sobre los datos de entrada, utilizando una función que toma las entradas de la capa anterior y las convierte en una nueva representación para la siguiente capa. Cada capa está definida por un conjunto de parámetros, como los pesos y sesgos, que permiten a la red aprender patrones complejos. Las capas procesan la información de manera conjunta, y a medida que los datos avanzan por cada capa, se transforman progresivamente hasta llegar a la capa final, donde se produce el resultado deseado, ya sea una predicción continua o una clasificación.
Existen tres tipos principales de capas en una red neuronal:
+ **Capa de entrada:** Recibe los datos iniciales del modelo, con cada neurona representando una característica del conjunto de datos.
+	**Capas ocultas:** Son las intermedias y realizan cálculos más complejos ajustando los pesos y aplicando funciones de activación para generar representaciones intermedias.
+ **Capa de salida:** Produce el resultado final, que puede ser una predicción continua en el caso de la regresión o una categoría en el caso de la clasificación.

6. **Normalizacion por lotes:** La normalización por lotes (Batch Normalization) es una técnica que busca reducir el "cambio covariante interno" en redes neuronales profundas. Este fenómeno ocurre cuando los cambios en los parámetros de una capa afectan la distribución de las entradas en las capas superiores, dificultando el entrenamiento del modelo. La normalización por lotes aborda este problema estandarizando las activaciones en cada capa, ajustándolas en función de la media y la desviación estándar calculadas a partir de un mini-lote de datos. Durante el entrenamiento, la normalización se realiza con las estadísticas de cada mini-lote, lo que permite que el modelo aprenda de manera más eficiente a través del descenso de gradiente. Durante la inferencia, las estadísticas se estiman con el conjunto completo de entrenamiento, lo que proporciona predicciones determinísticas.
7. **Regularizacion por dropout:** Dropout es una técnica de regularización utilizada en redes neuronales profundas para mitigar el sobreajuste (overfitting), que ocurre cuando un modelo se ajusta demasiado a los datos de entrenamiento, lo que compromete su capacidad de generalizar a nuevos datos. Dropout resuelve este problema desactivando de manera aleatoria algunas neuronas (o sus conexiones) durante el entrenamiento, lo que equivale a entrenar múltiples redes neuronales más pequeñas y variadas dentro de una red completa, sin tener que entrenarlas por separado. Al desactivar neuronas, se obliga a las restantes a volverse más robustas, promoviendo su independencia y capacidad para aprender características útiles sin depender excesivamente de otras. En la fase de prueba, se usa la red completa, pero los pesos se ajustan para compensar las neuronas inactivas durante el entrenamiento, ayudando a reducir el error de generalización.
5.	**Salida:** La salida es el resultado final que genera la neurona después de aplicar la función de activación. En un perceptrón simple, la salida es binaria, es decir, puede ser 0 o 1, lo que se interpreta como una clasificación en una de dos clases posibles.

**Modelo de Regresión**

Para aplicaciones de regresión, la estructuctura básica de la red neuronal profunda difiere respecto a una red hecha para clasificación, principalmente en la capa de salida y la función de activación utilizada. El proceso por el cual el modelo puede predecir una salida constante con las características de entrada es el siguiente:

Inicia con la capa de entrada donde se reciben las características de los datos de forma numérica, después se hace una sumatoria de la multiplicación de estas entradas por un peso aprendido y se les suma un sesgo en cada neurona. La ecuación que lo expresa es la siguiente:

$$z = \sum(w_i * x_i) + b$$

Donde W son los pesos aprendidos por la red, x son las entradas y b es el sesgo de esa nurona.

Una vez se tenga el valor de z de cada neurona se utiliza una función de activación que introduzca una no linealidad al modelo y permita capturar patrones complejos en los datos de entrada, las funciones mas comunes en capas ocultas son ReLU o Tanh. Este cálculo y activación se repite a lo largo de las capas ocultas hasta llegar a la salida, a esto se le conoce como propagación hacia adelante. Por último, la capa de salida predice un valor contínuo que suele ser simplemente el valor de z sin transformación no lineal.

Para la función de pérdida en problemas de regresión se usa comunmente el Error cuadrático medio o MSE para medir la diferencia entre el valor real y el valor predicho por el modelo. La ecuación es la siguiente:

$$L(y,\hat{y}) = \frac{1}{n} \sum_{i=1}^{n}(y_i - \hat{y}_i)^2$$

Donde y es el valor real, y hat es el valor predicho por el modelo y n es número de muestras. Con esta función se penalizan los errores grandes y el modelo se ajusta en el entrenamiento para minimizarlos.

Una vez que se calcula el error o pérdida, el modelo ajusta los pesos y los sesgos de todas las neuronas utilizando un proceso llamado backpropagation, en esta retropropagación el error calculado en la función de pérdida se retropropaga a través de la red, desde la capa de salida hasta la capa de entrada con la finalidad de calcular cómo cada peso y sesgo afectan el error final.

En este proceso se usan algoritmos de optimización como gradiente descendente o Adam para que el modelo ajuste los pesos y sesgos en cada neurona buscando minimizar el error. Esto ocurre en cada iteración o "época" durante el entrenamiento.

Ya que el modelo termino de entrenar y se tienen los pesos actualizados, se hace un proceso de predicción en nuevos datos de la misma forma que se hizo el cálculo hacia adelante, el modelo genereará una predicción con las nuevas características de entrada y se comparará el valor real con el predicho por el modelo para conocer su desempeño en la tarea de regresión. (SCCE, 2023)


**Modelo de Clasificación**

Cuando se trata de clasificación, el modelo utiliza un límite de decisión para asignar las entradas en distintas categorías o clases, para esto el límite de decisión actúa como una superficie que divide el espacio de las entradas en regiones donde cada una corresponde a una categoría. Cuando se recibe una nueva entrada, el modelo decide a qué región pertenece esta basado en su posición relativa al límite de decisión

Así como en regresión, la red neuronal profunda calcula una suma ponderada de un vector X como entrada de la siguiente forma:

$$z = \sum(w_i * x_i) + b$$

Donde W son los pesos aprendidos por la red, x son las entradas y b es el sesgo de esa nurona.

A continuación se utilizan funciones de activación como ReLU o Tanh en las capas ocultas para introducir una no linealidad al modelo, estas capas permiten construir límites de decisión complejos. El proceso de cálculo y activación se repite en cada capa hasta llegar a la capa de salida, a esto se le conoce como propagación hacia adelante.

En la capa de salida se aplica comunmente la función de activación Softmax. Esta convierte las salidas de la capa final (z) en probabilidades para cada clase que sumadas da un total de 1, esto es útil ya que en clasificación buscamos que la red calcule la probabilidad de que una entrada pertenezca a cada clase y seleccione la más probable. La ecuación de probabilidad de la función Softmax es la sguiente:

$$P(x_i) = \frac{e^{z_i}}{\sum_{j=1}^{k}} e^{z_j}$$

En la etapa de entrenamiento, la red ajusta sus pesos y sesgos para aprender un límite de decisión que separe las clases de manera óptima. Para ajustar estos pesos la red usa una función de pérdida para medir que tan lejos está la predicción hecha por el modelo con la clase real, comúnmente se usa Categorical Cross-Entropy como función de pérdida ya que esta mide la diferencia entre la distribución de probabilidades con la distribución real. La ecuación de esta función es:

$$L(y,\hat{y}) = - \sum_{i=1}^{n}y_i log(\hat{y}_i)$$

Donde y es el valor real, y hat es el valor predicho por el modelo y n es número de muestras. Para finalizar, al igual que en regresión se realiza la backpropagation para calcular los gradientes que muestran cómo ajustar los pesos en búsqueda de disminuir el error, para el ajuste de pesos y sesgos se usa un algoritmo de optimización como gradiente descendente o Adam, con el ajuste de estos pesos se modifica también los límites de decisión de manera que se separen con mayor eficiencia clases en el espacio.

### 2. Activation functions

En las redes neuronales, la función de activación es una función matemática que determina la salida de una neurona basada en su entrada. Su principal propósito es introducir una no linealidad al modelo, haciéndolo capaz de representar relaciones complejas en los datos.

- Activación lineal
La función de activación lineal es sencilla, ya que la salida es proporcional a la entrada sin aplicar ninguna transformación no lineal. Esta función se utiliza principalmente en problemas de regresión, donde se necesitan predicciones continuas. Sin embargo, no permite que la red neuronal capture relaciones no lineales entre las entradas y las salidas, lo que limita su capacidad en tareas más complejas. Es representada de la siguiente manera:

$$ F(x) = x$$

*Figura 2. Función de activación lineal*

- ReLu (Rectified Linear Unit)
La función de activación ReLU se caracteriza por dejar pasar los valores positivos tal como entran al sistema y anular los valores negativos. En términos simples, ReLU activa una neurona solo si su valor es positivo, mientras que, si el valor es negativo, la salida de la neurona será cero. ReLU es ampliamente utilizada en redes neuronales convolucionales (CNN), especialmente en tareas de procesamiento de imágenes, debido a su capacidad para mejorar el rendimiento. Sin embargo, una desventaja es que algunas neuronas pueden "morir", es decir, quedar inactivas permanentemente si reciben valores negativos durante el entrenamiento.

$$f(x) = \max(0,x)=\begin{cases} 0 &\text{for}x<0\\ x & \text{for} x \geq 0 \end{cases}$$

*Figura 3. Funcion de activación ReLu*


- Sigmoide
La función sigmoidal o logística convierte las entradas en un valor entre 0 y 1. Es común en problemas de clasificación binaria, ya que la salida puede interpretarse como una probabilidad. Sin embargo, en redes profundas puede sufrir del problema del gradiente descendiente, lo que dificulta el entrenamiento de capas más profundas. A continuación, se muestra la formula:

$$ Sigmoid (I_0)=\frac{1}{1+e^{-I_0}}$$

*Figura 4. Funcion de activación sigmoidal*

- Tanh
La función Tanh (tangente hiperbólica) es similar a la sigmoidal, pero en lugar de comprimir los valores entre 0 y 1, los ajusta en un rango entre -1 y 1 . Esta característica hace que las salidas estén centradas en cero, lo que puede mejorar el rendimiento en algunos modelos en comparación con la función sigmoidal. A pesar de ser una función no lineal, también puede sufrir del gradiente descendiente en redes profundas. A continuación, se muestra la fórmula de cómo se representa:

$$Tanh(I_0)=\frac{1-e^{-I_0}}{1+e^{I_0}}$$

*Figura 5. Funcion de activación Tanh*

- Softmax
La función de activación SoftMax es una técnica utilizada en problemas de clasificación multiclase, donde convierte las puntuaciones de las diferentes categorías en probabilidades interpretables. Dado un conjunto de puntuaciones obtenidas a partir de una combinación lineal de las características de entrada, la función SoftMax normaliza estos valores para que sumen 1, interpretándolos como probabilidades. A continuación, se muestra la formula con la que se representa:

$$ P(k) = \frac{e^{s_k}}{\sum_{j=1}^{K}e^{s_j}}$$

*Figura 6. Funcion de activación Softmax*

```python
# Entrenamiento del modelo
history8 = model8.fit(X_train, y_train,
                      epochs=100,
                      batch_size=16,
                      validation_split=0.2,
                      callbacks=[tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)])
```

### 5. Performance Metrics

```python
# Lista con los modelos entrenados
models = [model1, model2, model3, model4, model5, model6, model7, model8]

# Lista con los nombres de los modelos para identificar cada evaluación
model_names = ['Modelo 1', 'Modelo 2', 'Modelo 3', 'Modelo 4', 'Modelo 5', 'Modelo 6', 'Modelo 7', 'Modelo 8']

# Recorrer la lista de modelos entrenados
for i, model in enumerate(models):
    print(f"Evaluación del {model_names[i]}:")

    # Realiza predicciones
    y_pred = (model.predict(X_test) > 0.3).astype("int32")

    # Calcula la matriz de confusión
    cm = confusion_matrix(y_test.argmax(axis=1), y_pred.argmax(axis=1))  # Para multiclase con one-hot encoding

    # Calcula las métricas
    accuracy = accuracy_score(y_test.argmax(axis=1), y_pred.argmax(axis=1))
    precision = precision_score(y_test.argmax(axis=1), y_pred.argmax(axis=1), average='macro', zero_division=1)
    recall = recall_score(y_test.argmax(axis=1), y_pred.argmax(axis=1), average='macro', zero_division=1)
    f1 = f1_score(y_test.argmax(axis=1), y_pred.argmax(axis=1), average='macro', zero_division=1)


    # Especificidad por clase en multiclase
    specificity_per_class = []
    for j in range(len(cm)):
        tn = np.sum(cm) - (np.sum(cm[j, :]) + np.sum(cm[:, j]) - cm[j, j])
        fp = np.sum(cm[:, j]) - cm[j, j]
        specificity = tn / (tn + fp) if (tn + fp) != 0 else 0
        specificity_per_class.append(specificity)

    # Especificidad promedio
    specificity_avg = np.mean(specificity_per_class)

    # Imprimir las métricas
    print(f'Accuracy: {accuracy}')
    print(f'Precision: {precision}')
    print(f'Recall: {recall}')
    print(f'Specificity (Average): {specificity_avg}')
    print(f'F1-Score: {f1}')

    # Generar el gráfico de la matriz de confusión
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f'Matriz de Confusión - {model_names[i]}')
    plt.show()
```
